```python
import numpy as np
marks=np.array([35,40,48,55,70])
total_marks=np.sum(marks)
print(total_marks)
average_marks=np.mean(marks)
print(average_marks)
min_marks=np.min(marks)
max_marks=np.max(marks)
print(min_marks,"\n",max_marks)
standird_deviation=np.std(marks)
print(standird_deviation)
```

```python
import numpy as np

# Original monthly salaries
salaries = np.array([25000, 30000, 40000, 50000])

# 1. Increase each salary by 10% (element-wise operation)
increased_salaries = salaries * 1.10

# 2. Calculate yearly salary (monthly × 12)
yearly_salaries = increased_salaries * 12

# 3. Find employees earning more than 35,000 (boolean indexing)
high_earners = increased_salaries[increased_salaries > 35000]

print("Original Salaries:", salaries)
print("Salaries after 10% increase:", increased_salaries)
print("Yearly Salaries:", yearly_salaries)
print("Employees earning more than 35,000:", high_earners)
```

```python
import numpy as np

# Generate 12 random integers between 1 and 100
data = np.random.randint(1, 101, 12)

# Reshape into a 3×4 matrix
matrix = data.reshape(3, 4)

print("3x4 Matrix:")
print(matrix)

# Print each row separately (iteration)
print("\nEach row:")
for row in matrix:
    print(row)

# Flatten the matrix back to 1D
flattened = matrix.flatten()
print("\nFlattened array:")
print(flattened)

# Find sum of each row
row_sums = np.sum(matrix,axis=1)
print("\nSum of each row:")
print(row_sums)
```
